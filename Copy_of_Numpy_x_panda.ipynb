```python
import matplotlib.pyplot as plt

# Line chart
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o', color='blue')
plt.title("Line Chart")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
# Bar chart
subjects = ['Math', 'Science', 'English']
marks = [80, 90, 70]

plt.bar(subjects, marks, color=['blue', 'green', 'orange'])
plt.title("Marks by Subject")
plt.ylabel("Marks")
plt.show()
```

```python
import numpy as np
#Histogram
data = np.random.normal(60, 70, 50)
plt.hist(data, bins= 50, color='pink')
plt.title("Histogram")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Scatter plot
x = np.random.rand(30)
y = np.random.rand(30)

plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.show()
```

```python
# Pie chart
labels = ['Mobile', 'Desktop', 'Tablet']
values = [50, 35, 15]

plt.pie(values, labels=labels, autopct='%1.0f%%')
plt.title("Device Usage")
plt.show()

print("✅ Day 1 done!")
```
